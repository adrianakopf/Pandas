```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()

```

```python
purchase_data = purchase_data.rename(columns={"SN": "Players"})
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
dropped_players = purchase_data.drop_duplicates(["Players"], keep="last")
players_count = dropped_players["Players"].count()
print(players_count)

```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
#number of items
dropped_items = purchase_data.drop_duplicates(["Item ID"], keep="last")
items_count = dropped_items["Item ID"].count()
print(items_count)
```

```python
#average price
item_average = dropped_items["Price"].mean()
print(item_average)
```

```python
#number of purchases
number_purchases = purchase_data["Purchase ID"].count()
print(number_purchases)
```

```python
total_revenue = item_average * number_purchases
print(total_revenue)
```

```python
#display clean data formatting 
#hold the result, show
item_summary_table = pd.DataFrame({
    "Number of Unique Items": [items_count],
    "Number of Purchases": [number_purchases], 
    "Average Price": (item_average),
    "Total Revenue": (total_revenue),
})
item_summary_table["Total Revenue"] = item_summary_table["Total Revenue"].map("${:.2f}".format)
item_summary_table["Average Price"] = item_summary_table["Average Price"].map("${:.2f}".format)
item_summary_table


```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Establish bins for ages
bins = [0,10,20,30,40,50]
group_labels = ["Under 10", "11-20", "21-30", "31-40", "41-50"]
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)
purchase_data.head()

#Purchasing Analysis (Gender)

age_data_total = purchase_data["Age Range"].value_counts()
purchase_data_per = (age_data_total/players_count) * 100

age_demographics_table = pd.DataFrame({"Total Count": age_data_total, "Percentage of Players": purchase_data_per })
age_demographics_table = age_demographics_table.sort_index()

age_demographics_table["Percentage of Players"] = age_demographics_table["Percentage of Players"].map("{:.2f}%".format)
age_demographics_table
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#obtain purchase count, avg. purchase price, avg. purchase total per person
group_bin = purchase_data.groupby(["Age Range"])
range_purchase_count = group_bin["Players"].count()
age_avg_price = group_bin["Price"].mean()
age_total_purchase = group_bin["Price"].sum()
avg_per_person = age_total_purchase / range_purchase_count

age_purchase_data = pd.DataFrame({"Purchase Count": range_purchase_count, 
                            "Average Purchase Price": age_avg_price, 
                            "Total Purchase Value": age_total_purchase, 
                            "Avg Total Purchase per Person": avg_per_person})

age_purchase_data["Average Purchase Price"] = age_purchase_data["Average Purchase Price"].map("${:.2f}".format)
age_purchase_data["Total Purchase Value"] = age_purchase_data["Total Purchase Value"].map("${:.2f}".format)
age_purchase_data["Avg Total Purchase per Person"] = age_purchase_data["Avg Total Purchase per Person"].map("${:.2f}".format)
age_purchase_data = age_purchase_data[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]
age_purchase_data
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spender = purchase_data.groupby("Players")["Price"]
top_spender_count = top_spender.count()
top_spender_total = top_spender.sum()
top_spender_avg = top_spender.mean()

top_spender_data = pd.DataFrame({"Purchase Count": top_spender_count,
                          "Average Purchase Price": top_spender_avg,
                          "Total Purchase Value": top_spender_total})

top_spender_data["Average Purchase Price"] = top_spender_data["Average Purchase Price"].map("${:.2f}".format)
top_spender_data["Total Purchase Value"] = top_spender_data["Total Purchase Value"].map("${:.2f}".format)

top_spender_data.sort_values("Total Purchase Value", ascending=False).head(10)
```

## Most Popular Items

* Create a summary data frame to hold the results

* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Most Popular Items
item_df = purchase_data.groupby("Item ID")["Price"]
item_count = item_df.count()
item_total = item_df.sum()
item_price = item_df.mean()
item_name = purchase_data["Item Name"]
item_id = purchase_data["Item ID"]

item_data = pd.DataFrame({"Item ID": item_id, 
                          "Item Name": item_name, 
                          "Purchase Count": item_count,
                          "Item Price": item_price,
                          "Total Purchase Value": item_total})
item_data["Item Price"] = item_data["Item Price"].map("${:.2f}".format)
item_data["Total Purchase Value"] = item_data["Total Purchase Value"].map("${:.2f}".format)

most_popular_item = item_data.sort_values(by=["Purchase Count"], ascending=False)
most_popular_item.head(5)

```
